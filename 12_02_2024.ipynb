```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/1000_Companies.csv')
df
```

```python
#dependent and independent variabless identification
x=df.drop('Profit',axis='columns')
x
```

```python
y = df['Profit']
y
```

step-3:heatmap(corelation)

```python
sns.heatmap(df.corr())
```

step-4&5:identify categorical data and encoding

```python
dummies=pd.get_dummies(df['State'])
dummies
```

```python
merged = pd.concat([x,dummies], axis='columns')
merged
```

```python
final = merged.drop(['State', 'California'], axis='columns')
final
```

step-6:split data

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(final,y,test_size=0.6)
```

```python
len(x_train)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

step-7:model training

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

step-8:predicting the values

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

step-9:model metric

```python
model.score(x_test,y_test)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('/content/insurance_data.csv')
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance,marker='+',color='red')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df[['age']],df.bought_insurance,train_size=0.2)
```

```python
x_test
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_test
```

```python
model.score(x_test,y_test)
```
